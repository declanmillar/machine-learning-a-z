# Support Vector Machine

Support Vector Machines (SVMs) were initially developed in the 1960s and refined in the 1990s, but are becoming very popular now as they are demonstrating can be quite powerful, as they are quite different to other techniques.

So, how do they work? Let's assume we have some pre-classified data, as shown below.



Assuming we are to introduce new points later, how could we classfiy these? We might expect to derive a line to separate these classes. 

Naively, we might try to draw a straight vertical line in the 2D space, like so:



Any new points we simply be classifed by whether it is on the left or right. We could, equally arbitrarily, have chosen a horizontal or any diagonal line.



Lots of different lines could be chosen to classify the data, which would have different consequences for the classification of new data. SVMs select for the best line or, more generally, decision boundary to classify new data according to classified data. They are therefore a type of supervised machine learning.


## Maximum margin

We wish to maximise the margins between the decision surface and the classes of points, such that the line is equidistant between them.




## Support vectors

The points that will ultimately lie cloest to this line are known as the support vectors. They support the whole algorithm that derives this line. The rest of points do not contribute to the algorithm; even if they were all removed, the algorithm would be unchanged. 



While here we address these data points as points, they are of course 2D vectors.



When the number of variables exceeds three, one can no longer comfortably visualise them as points, yet vector calculus remains valid. The decision surface comprises the maximum margin hyperplane (classifier).



The dotted lines through the support vectors are known as the positive or negative hyperplanes.



## Why are SVMs special?

SVMs are special because they classify based on the outliers (support vectors) of the classes, rather than the proto-typical examples in the heart of the samples.
