# Kernel SVM

## Motivation

When trying to apply an SVM to a classification problem, what happens when you can't find a hyperplane that separates the classes (see image).



This has happened because, in this case, the data points are not *linearly separable*. SVMs allow us to find the optimum decision boundary, on the assumption of linear separability. However, in this case, we cannot draw any suitable hyperplanes. How might we overcome this?



## Mapping to a higher dimensional space

We're now going to see how we can map the non-linearly separable space to a linearly separable one for which we can apply a SVM. Let's take the simplest example of a 1-dimensional model, with a class of points surrounded on both sides by a different class of points (see image).



The decision hyperplane in a problem of this dimensionality would be a 0-dimensional scalar. There is nowhere that we can place such a boundary that would separate the classes.
